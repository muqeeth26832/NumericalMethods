```python
import numpy as np
import matplotlib.pyplot as plt

def gauss_legendre_quadrature(n):
    J = np.zeros((n, n)) 
    for i in range(1, n):
        J[i, i - 1] = J[i - 1, i] = np.sqrt(i * i / ((4 * i**2) - 1))

    eigenvalues, eigenvectors = np.linalg.eigh(J)

    # Roots are eigenvalues
    roots = eigenvalues

    # Weights are square of the first element of each eigenvector (Golub-Welsch Theorem)
    weights = 2 * (eigenvectors[0, :] ** 2)

    return roots, weights
  
# Given n
n = 64 
roots, weights = gauss_legendre_quadrature(n)

print("Roots (nodes):", roots)
print("Weights:", weights)

# Plotting weights against the roots
plt.figure(figsize=(10, 6))
plt.plot(roots, weights, 'bo-', label="Quadrature Weights")
plt.xlabel("Roots (Nodes)")
plt.ylabel("Weights")
plt.title("Weights vs. Roots for 64th-order Legendre Polynomial (Gaussian Quadrature)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#PartB
import numpy as np
from numpy.polynomial import Polynomial
from scipy.integrate import quad
import matplotlib.pyplot as plt

# Function to compute the Lagrange polynomial
def lagrange_poly(x, i, roots_b):
    L_i = 1
    for j in range(len(roots_b)):
        if i != j:
            L_i *= (x - roots_b[j]) / (roots_b[i] - roots_b[j])
    return L_i

# Function to generate the Legendre polynomial of order n
def legendre_polynomial(n):
    # Initialize base polynomials P_0(x) = 1 and P_1(x) = x
    P0 = Polynomial([1])        # P_0(x) = 1
    P1 = Polynomial([0, 1])     # P_1(x) = x

    for k in range(1, n):
        P_next = ((2 * k + 1) * Polynomial([0, 1]) * P1 - k * P0) / (k + 1)
        P0, P1 = P1, P_next

    return P1


n = int(input("Enter the order of the Legendre polynomial : "))
Pn = legendre_polynomial(n)

# Finding the coefficients for the companion matrix
coefficients = Pn.coef[::-1]  # Reverse order

# Creating the companion matrix
companion_matrix = np.diag(np.ones(n - 1), -1)
companion_matrix[0, :] = -coefficients[1:] / coefficients[0]

# Finding eigenvalues (roots)
roots = np.sort(np.linalg.eigvals(companion_matrix))

print("Roots of the", n, "th-order Legendre polynomial:\n", roots)

# Finding weights by integrating the Lagrange polynomials
weights = []
for i in range(len(roots)):
    # Integrate L_i(x) from -1 to 1 to compute weight w_i
    weight_i, _ = quad(lagrange_poly, -1, 1, args=(i, roots))
    weights.append(weight_i)

weights = np.array(weights)

print("\nWeights:\n", weights)

print("companion_matrix:\n", companion_matrix)
# Plotting weights against the roots
plt.figure(figsize=(10, 6))
plt.plot(roots, weights, 'bo-', label="Quadrature Weights")
plt.xlabel("Roots")
plt.ylabel("Weights")
plt.title(f"Weights vs. Roots for {n}-th order Legendre Polynomial (Gaussian Quadrature)")
plt.legend()
plt.grid(True)
plt.show()
```
